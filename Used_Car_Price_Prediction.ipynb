```python
#outlier removal(mpg values of Electric,Hybrid and Other are extreme)
df6=df5.drop(df5[df5.fuelType=='Electric'].index,axis='rows')
df6=df6.drop(df5[df5.fuelType=='Other'].index,axis='rows')
df6=df6.drop(df5[df5.fuelType=='Hybrid'].index,axis='rows')
df6.groupby('fuelType')['fuelType'].agg('count')
```

```python
df7=df6.copy()
df7.fuelType=df6.fuelType.apply(lambda x: 1 if x=='Diesel' else 0)
df7
```

```python
dummies1=pd.get_dummies(df7.Make,drop_first='true')
dummies1
```

```python
df8 = pd.concat([df7.drop('Make',axis='columns'),dummies1],axis='columns')
df8
```

```python
dummies2=pd.get_dummies(df8.model,drop_first='true')
dummies2
```

```python
df9=pd.concat([df8.drop('model',axis='columns'),dummies2],axis='columns')
df9
```

```python
X=df9.drop('price',axis='columns')
X
```

```python
y=df9.price
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train.values,y_train.values)
lr_clf.score(X_test.values,y_test.values)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2)

cross_val_score(LinearRegression(), X, y, cv=cv).mean()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
clf = GridSearchCV(LinearRegression(), {
    'fit_intercept': [True, False]
}, cv=5, return_train_score=False)
clf.fit(X, y)
clf.cv_results_
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
clf = GridSearchCV(DecisionTreeRegressor(), {
     'criterion' : ['squared_error','friedman_mse'],
     'splitter': ['best','random']
}, cv=5, return_train_score=False)
clf.fit(X, y)
clf.cv_results_
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
#Decision_tree has the best_score with friedman_mse and best for criterion and splitter respectively
dcn_tr_clf=DecisionTreeRegressor()
Xnp=X.columns.to_numpy()
Xnp
```

```python
def predict_price(model,Make,year,transmission,mileage,fuelType,engineSize):    
    model_index = np.where(Xnp==model)
    Make_index=np.where(Xnp==Make)
    if(transmission=='Manual'):
        tr=0
    else:
        tr=1
    if(fuelType=='Petrol'):
        ft=0
    else:
        ft=1
    x = np.zeros(len(X.columns))
    x[0] = year
    x[1] = tr 
    x[2] = mileage
    x[3] = ft 
    x[4] = engineSize
   
    x[model_index] = 1
    x[Make_index] = 1

    return lr_clf.predict([x])[0]
```
